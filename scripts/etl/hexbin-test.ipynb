```python
import pandas as pd

# 1. Carica il dataset
# Assicurati che il file sia nella stessa cartella dello script 
# o fornisci il perc\orso completo (es. 'C:/Utenti/Desktop/file.csv')
nome_file = 'children-out-school.csv' 
df = pd.read_csv('../../datasets/raw/children-out-school/children-out-school.csv')

# 2. Definisci le regioni da filtrare
regioni_target = ['Northern Africa and Western Asia', 'Sub-Saharan Africa']
categoria_target = 'Region'

# 3. Filtra il dataset con entrambe le condizioni
# Nota: quando si usano più filtri in pandas, ogni condizione va tra parentesi tonda ()
df_filtrato = df[
    (df['region_group'].isin(regioni_target)) & 
    (df['category'] == categoria_target)
]

# 4. (Opzionale) Visualizza i risultati
print(f"Righe totali originali: {len(df)}")
print(f"Righe dopo il filtraggio: {len(df_filtrato)}")
print(df_filtrato.head())

# 5. (Opzionale) Salva il file filtrato in un nuovo CSV
df_filtrato.to_csv('dataset_africa_asia.csv', index=False)
```

```python
df_africa = pd.read_csv('dataset_africa_asia.csv')
df_continenti = pd.read_csv('../../datasets/raw/country-codes/country-code.csv')

df_continenti_unique = df_continenti.drop_duplicates(subset=['Three_Letter_Country_Code'])

# 2. Esegui la JOIN (Merge)
# left_on e right_on si usano quando le colonne chiave hanno nomi diversi
df_unito = pd.merge(
    df_africa, 
    df_continenti_unique[['Three_Letter_Country_Code', 'Continent_Name']], 
    left_on='iso_code', 
    right_on='Three_Letter_Country_Code', 
    how='left'
)

df_finale = df_unito[df_unito['Continent_Name'] == 'Africa'].copy()

# Pulizia: rimuoviamo la colonna doppione della join
df_finale.drop(columns=['Three_Letter_Country_Code'], inplace=True)
print(f"Righe dopo il filtraggio: {len(df_finale)}")

df_finale.to_csv('dataset_africa.csv', index=False)

```
